```python
test_values=pd.read_csv('/content/test_values.csv')
```

```python
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

for col in train_values.columns:
    if train_values[col].dtype == 'object':
        # Fit label encoder and transform values
        train_values[col] = label_encoder.fit_transform(train_values[col])
```

```python
for col in test_values.columns:
    if test_values[col].dtype == 'object':
        # Fit label encoder and transform values
        test_values[col] = label_encoder.fit_transform(test_values[col])
```

```python
predictors= ['geo_level_1_id', 'geo_level_2_id', 'geo_level_3_id','count_floors_pre_eq', 'age','area_percentage','height_percentage','land_surface_condition','foundation_type','roof_type','ground_floor_type','other_floor_type','position',
       'plan_configuration','has_superstructure_adobe_mud',
       'has_superstructure_mud_mortar_stone', 'has_superstructure_stone_flag','has_superstructure_cement_mortar_stone','has_superstructure_mud_mortar_brick',
       'has_superstructure_cement_mortar_brick', 'has_superstructure_timber',
       'has_superstructure_bamboo','has_superstructure_rc_non_engineered',
       'has_superstructure_rc_engineered', 'has_superstructure_other']
```

```python
target=['damage_grade']
```

```python
train_values=train_values[predictors]
```

```python
train_values=pd.get_dummies(train_values)
```

```python
from sklearn.model_selection import train_test_split

X_train = train_values[predictors]
y_train = train_labels[target]

X_test = test_values[predictors]
```

```python
from sklearn.preprocessing import StandardScaler

# Instantiate the scaler
scaler = StandardScaler()

# Fit the scaler to the training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test/validation data using the same scaler
X_test_scaled=scaler.transform(X_test)
```

```python
X_test_encoded = pd.get_dummies(X_test)

# Scale the encoded test data
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
X_train_scaled.shape
```

```python
y_train_encoded = y_train - 1
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
```

```python
from xgboost import XGBClassifier
model_xgb=XGBClassifier()
model_xgb.fit(X_train_scaled, y_train_encoded)
```

```python
in_sample_test=model_xgb.predict(X_test_scaled)
```

```python
from sklearn.metrics import f1_score

# Assuming y_train contains the true labels for your training data
# If y_train is a DataFrame, you may need to extract the relevant column containing the labels
# For example, if 'damage_grade' is the column containing labels, you can do y_train['damage_grade']

# Compute the predictions using the pipeline
in_sample_preds = model_xgb.predict(X_train_scaled)

# Calculate the F1 score
f1 = f1_score(y_train_encoded, in_sample_preds, average='micro')

print("F1 Score for training data:", f1)
```
