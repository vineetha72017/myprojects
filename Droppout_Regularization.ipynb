```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/sonar_dataset.csv',header=None)
df.head()
```

```python
df.shape
```

```python
df.isna().sum().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
X=df.drop(60,axis=1)
y=df[60]
```

```python
y
```

```python
y = pd.get_dummies(y, drop_first=True)
y
```

```python
y.value_counts()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')

])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test).reshape(-1)
y_pred[:10]
```

```python
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```

```python
modelid=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.60),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.55),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.55),
    keras.layers.Dense(1,activation='sigmoid')

])
modelid.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
modelid.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
modelid.evaluate(X_test,y_test)
```

```python
y_pred=modelid.predict(X_test).reshape(-1)
y_pred
```

```python
y_pred=np.round(y_pred)
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```
