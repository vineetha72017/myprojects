Do Label encoding of object dtypes of train labels

```python
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
train_values['land_surface_condition']= label_encoder.fit_transform(train_values['land_surface_condition'])

train_values['land_surface_condition'].unique()
```

```python
for col in train_values.columns:
    if train_values[col].dtype == 'object':
        # Fit label encoder and transform values
        train_values[col] = label_encoder.fit_transform(train_values[col])

```

Do Label Encoding of object type columns of test values

```python
for col in test_values.columns:
    if test_values[col].dtype == 'object':
        # Fit label encoder and transform values
        test_values[col] = label_encoder.fit_transform(test_values[col])

```

```python
predictors=['foundation_type','area_percentage','height_percentage','count_floors_pre_eq','land_surface_condition','has_superstructure_cement_mortar_stone']
```

```python
target=['damage_grade']
```

```python
train_values.columns
```

```python
from sklearn.model_selection import train_test_split

X_train = train_values[predictors]
y_train = train_labels[target]

X_test = test_values[predictors]

# Splitting the data into train and test sets
```

```python
train_labels['damage_grade'].value_counts()
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

# Instantiate the scaler
scaler = StandardScaler()

# Fit the scaler to the training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test/validation data using the same scaler
X_test_scaled = scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(25, input_shape=(X_train.shape[1],), activation='relu')) # input shape is (features,)
model.add(Dense(6, activation='softmax'))
model.summary()

# compile the model
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy', # this is different instead of binary_crossentropy (for regular classification)
              metrics=['accuracy'])
```

```python
from keras.utils import to_categorical

# Assuming y_train contains integer labels representing classes
num_classes = 6  # Number of classes in your classification problem

# Convert target labels to one-hot encoded format
y_train_encoded = to_categorical(y_train, num_classes=num_classes)

# Now y_train_encoded should have the shape (number of samples, number of classes)

# Train the model
```

```python
history = model.fit(X_train_scaled, y_train_encoded,
                    epochs=100,
                    batch_size=10,
                    shuffle=True,
                    validation_split=0.2,
                    verbose=1)
```

```python
from sklearn.metrics import f1_score
in_sample_preds=model.predict(X_train_scaled)
from sklearn.metrics import f1_score

# Assuming in_sample_preds contains the predicted labels for your training data
# You may need to convert the predicted probabilities to class labels depending on your model output
# If your model outputs probabilities, you can use np.argmax to convert them to class labels
in_sample_preds_classes = np.argmax(in_sample_preds, axis=1)

# Assuming train_labels are the true labels for your training data
# Make sure train_labels are 1D array
# If not, you may need to flatten or reshape them
train_labels_series = train_labels['damage_grade']  # Assuming 'damage_grade' is the column containing the true labels

# Calculate the F1 score
f1 = f1_score(train_labels_series, in_sample_preds_classes, average='micro')
```
