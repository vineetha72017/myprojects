```python
from sklearn.ensemble import RandomForestClassifier
import category_encoders as ce 
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer

#build a pipeline for a ransom forest
pipeline = make_pipeline(
    ce.OneHotEncoder(use_cat_names=True), 
    SimpleImputer(strategy='median'), 
    RandomForestClassifier(n_jobs=-1, bootstrap = True, oob_score =True, random_state=25))
    #The number of jobs to run in parallel. fit, predict, 
    # decision_path and apply are all parallelized over the trees. None means 1 unless in a joblib.parallel_backend context. -1 means using all processors

#fit on train, and score on val 
pipeline.fit(X_train, y_train)
```
