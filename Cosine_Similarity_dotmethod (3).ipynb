```python
def compute_similarity1(df):
  
  #Defining two empty lists
  cos_sim_values = []
  
  
  #Run a loop to compute similarity between two vectors
  for i in range(0,len(df)):
      
      #storing elements from the cell in Product_name
      name = (df['name'][i:i+1])

      #converting to string value
      input_name = name.to_string()



      #storing elements from the cell in Product_url_clean
      screen_name = (df['screen_name'][i:i+1])

      #converting to string value
      input_screen_name = screen_name.to_string()

      # Removing any digit association with the text
      #screen_name = ''.join([m for m in screen_name if not m.isdigit()])
      
      description = (df['description'][i:i+1])

      #converting to string value
      input_description = description.to_string()

      # Removing any digit association with the text
      #description = ''.join([j for j in name if not j.isdigit()])


      #splitting the words and storing it in dataframe
      name_df = pd.DataFrame({"Words": input_name},index=[0])

      #lookup for the frequency of the words
      name_df_freq = pd.DataFrame(pd.crosstab(index=name_df['Words'],columns='count'))

      #splitting the words and storing it in dataframe
      screen_df = pd.DataFrame({"Words1": input_screen_name.split()})

      #lookup for the frequency of the words
      screen_df_freq = pd.DataFrame(pd.crosstab(index=screen_df['Words1'],columns='count'))

      description_df = pd.DataFrame({"Words2": input_description.split()})

      #lookup for the frequency of the words
      description_df_freq = pd.DataFrame(pd.crosstab(index=description_df['Words2'],columns='count'))

      #concat the dfs
      dfs = [name_df_freq, screen_df_freq]
      all_words = pd.concat(dfs, axis=1)
      dfs1=[name_df_freq,description_df_freq]
      all_words1 = pd.concat(dfs1, axis=1)
      dfs2=[description_df_freq,screen_df_freq]
      all_words2 = pd.concat(dfs2, axis=1)

      # Substituting the NAN with zeros
      all_words = all_words.fillna(0)
      all_words.columns = ["name_data", "screen_data"]
      all_words1= all_words1.fillna(0)
      all_words1.columns=["name_data","description_data"]
      all_words2= all_words2.fillna(0)
      all_words2.columns=["description_data","screen_data"]
      #computing the cosine values
      cos_sim = dot(all_words["name_data"], all_words["screen_data"])/(norm(all_words["name_data"])*norm(all_words["screen_data"]))
      cos_sim = dot(all_words1["name_data"], all_words1["description_data"])/(norm(all_words1["name_data"])*norm(all_words1["description_data"]))
      cos_sim = dot(all_words2["description_data"], all_words2["screen_data"])/(norm(all_words2['description_data'])*norm(all_words2["screen_data"].iloc[1]))

      #append the values in the corresponding lists
      cos_sim_values.append(cos_sim)

  #storing the above lists values in the newly created column        
  df['cos_sim_values'] = cos_sim_values

  #Return the value
  return df
```

```python
def compute_similarity3(df):
    cos_sim_name_values = []
    cos_sim_screen_values = []
    cos_sim_description_values = []

    for i, row in df.iterrows():
        name1 = str(row['name']) if not isinstance(row['name'], str) else row['name']
        screen_name1 = str(row['screen_name']) if not isinstance(row['screen_name'], str) else row['screen_name']
        description1 = str(row['description']) if not isinstance(row['description'], str) else row['description']

        cos_sim_name = []
        cos_sim_screen = []
        cos_sim_description = []

        for j, row2 in df.iterrows():
            name2 = str(row2['name']) if not isinstance(row2['name'], str) else row2['name']
            screen_name2 = str(row2['screen_name']) if not isinstance(row2['screen_name'], str) else row2['screen_name']
            description2 = str(row2['description']) if not isinstance(row2['description'], str) else row2['description']

            all_words_name = pd.DataFrame({"Words": [name1, name2]})
            all_words_name_freq = pd.DataFrame(pd.crosstab(index=all_words_name['Words'], columns='count'))

            all_words_screen = pd.DataFrame({"Words": [screen_name1, screen_name2]})
            all_words_screen_freq = pd.DataFrame(pd.crosstab(index=all_words_screen['Words'], columns='count'))

            all_words_description = pd.DataFrame({"Words": [description1, description2]})
            all_words_description_freq = pd.DataFrame(pd.crosstab(index=all_words_description['Words'], columns='count'))

            cos_sim_name_value = dot(all_words_name_freq['count'], all_words_name_freq['count']) / (norm(all_words_name_freq['count']) * norm(all_words_name_freq['count']))
            cos_sim_screen_value = dot(all_words_screen_freq['count'], all_words_screen_freq['count']) / (norm(all_words_screen_freq['count']) * norm(all_words_screen_freq['count']))
            cos_sim_description_value = dot(all_words_description_freq['count'], all_words_description_freq['count']) / (norm(all_words_description_freq['count']) * norm(all_words_description_freq['count']))

            cos_sim_name.append(cos_sim_name_value)
            cos_sim_screen.append(cos_sim_screen_value)
            cos_sim_description.append(cos_sim_description_value)

        cos_sim_name_values.append(cos_sim_name)
        cos_sim_screen_values.append(cos_sim_screen)
        cos_sim_description_values.append(cos_sim_description)

    df['cos_sim_name_values'] = cos_sim_name_values
    df['cos_sim_screen_values'] = cos_sim_screen_values
    df['cos_sim_description_values'] = cos_sim_description_values

    return df
```

```python
def compute_similarity2(df):
    # Defining an empty list
    cos_sim_values = []

    # Run a loop to compute similarity between rows
    for i in range(len(df)):
        # Extracting values from the columns
        name = df.loc[i, 'name']
        screen_name = df.loc[i, 'screen_name']
        description = df.loc[i, 'description']

        # Converting values to strings
        input_name = str(name)
        input_screen_name = str(screen_name)
        input_description = str(description)

        # Splitting the words and storing them in dataframes
        name_words = input_name.split()
        screen_words = input_screen_name.split()
        description_words = input_description.split()

        # Computing the unique words from the columns
        all_words = list(set(name_words + screen_words))
        all_words1 = list(set(name_words + description_words))
        all_words2 = list(set(description_words + screen_words))

        # Creating frequency dictionaries
        name_freq = {word: name_words.count(word) for word in all_words}
        screen_freq = {word: screen_words.count(word) for word in all_words}
        name_description_freq = {word: name_words.count(word) for word in all_words1}
        description_freq = {word: description_words.count(word) for word in all_words1}
        description_screen_freq = {word: description_words.count(word) for word in all_words2}
        screen_freq_2 = {word: screen_words.count(word) for word in all_words2}

        # Computing the cosine similarity values
        cos_sim_name_screen = dot(list(name_freq.values()), list(screen_freq.values())) / (
                norm(list(name_freq.values())) * norm(list(screen_freq.values())))
        cos_sim_name_description = dot(list(name_description_freq.values()), list(description_freq.values())) / (
                norm(list(name_description_freq.values())) * norm(list(description_freq.values())))
        cos_sim_description_screen = dot(list(description_screen_freq.values()), list(screen_freq_2.values())) / (
                norm(list(description_screen_freq.values())) * norm(list(screen_freq_2.values())))

        # Compute the average cosine similarity
        cos_sim = (cos_sim_name_screen + cos_sim_name_description + cos_sim_description_screen) / 3

        # Append the cosine similarity value to the list
        cos_sim_values.append(cos_sim)

    # Store the cosine similarity values in a new column
    df['cos_sim_values'] = cos_sim_values

    # Return the dataframe
    return df
```
